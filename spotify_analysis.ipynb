# Outliers Treatment using Capping Approach
   ### IQR Method to treat Outliers:

All the values below Q1 - 1.5*IQR and values above Q3 + 1.5*IQR are outliers and can be removed.


```python
# Finding the Quantiles
Q1 = sub_table_loc['spotify_listening_device'].quantile(0.25)  # First Quartile (25th percentile)
Q2 = sub_table_loc['spotify_listening_device'].quantile(0.50)  # Second Quartile (50th percentile, median)
Q3 = sub_table_loc['spotify_listening_device'].quantile(0.75)  # Third Quartile (75th percentile)

# IQR: Inter-Quartile Range
IQR = Q3 - Q1

# Lower Limit
LC = Q1 - (1.5 * IQR)

# Upper Limit
UC = Q3 + (1.5 * IQR)

# Display the lower and upper limits
print("Lower Limit:", LC)
print("Upper Limit:", UC)
```

```python
## Plot

sns.distplot(sub_table_loc.spotify_listening_device)
plt.axvline(UC, color='r')
plt.axvline(LC, color ='r')
plt.axvline(Q1, color='g')
plt.axvline(Q3, color='g')
plt.show()
```

```python
# Find count of Outliers wrt IQR

sub_table_loc[(sub_table_loc.spotify_listening_device<LC) | (sub_table_loc.spotify_listening_device>UC)].reset_index(drop=True)
```

```python


sub_table_loc[(sub_table_loc.spotify_listening_device<LC) | (sub_table_loc.spotify_listening_device>UC)].shape[0]
```

```python
## Store the clean data wrt IQR:

df3 = sub_table_loc[(sub_table_loc.spotify_listening_device>=LC) & (sub_table_loc.spotify_listening_device<=UC)]
```

# Linear Regression Model Building and Training

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
# Check shapes after split
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
# Fit the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
# Calculate metrics
mse_rf = mean_squared_error(y_test, y_pred)
r2_rf = r2_score(y_test, y_pred)
```

```python
# Evaluate the Linear Regression model
#from sklearn.metrics import mean_absolute_error as mas
y_pred = model.predict(X_test)
print('Linear Regression MAE:', mean_squared_error(y_test, y_pred))
print('Linear Regression R-squared:', r2_score(y_test, y_pred))
```

```python
# Metrics dictionary
metrics_rf = {
    'Mean Squared Error (MSE)': mse_rf,
    'R-squared': r2_rf
}
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have your data in a DataFrame named 'df'

# 1. Willingness vs. Preferred Premium Plan

# Create a cross-tabulation table
contingency_table = pd.crosstab(df['premium_sub_willingness'], df['preffered_premium_plan'])

# Get data for visualization (assuming 'Individual' and 'Family' are plan types)
willingness_categories = contingency_table.index  # Get willingness categories (e.g., "Yes", "No", "Unsure")
plan_types = contingency_table.columns  # Plan types (e.g., "Individual", "Family")
plan_counts_by_willingness = contingency_table.values  # Plan counts for each willingness category

# Create colors for bars (customize as desired)
willingness_colors = ['lightgreen', 'lightsteelblue', 'lightcoral']

# Create a stacked bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.35  # Adjust bar width as needed
index = range(len(plan_types))  # Create x-axis positions

# Stacked bars for plan counts within each willingness category
for i, willingness in enumerate(willingness_categories):
  plt.bar(index, plan_counts_by_willingness[i], bar_width, label=willingness, bottom=sum(plan_counts_by_willingness[:i]), color=willingness_colors[i])

# Label the chart
plt.xlabel("Premium Plan")
plt.ylabel("Count")
plt.title("Preferred Premium Plan by Spotify Subscription Willingness")
plt.xticks([i + bar_width / 2 for i in index], plan_types, rotation=45, ha='right')  # Rotate x-axis labels
plt.legend(title="Willingness")
plt.subplots_adjust(bottom=0.3)  # Adjust space for legend

# Display the stacked bar chart
plt.show()


```

# Premium Subscription Willingness

```python
# 2. Preferred Listening Content vs. Premium Willingness

# Group data by listening content
listening_content_groups = df.groupby('preferred_listening_content')

# Analyze willingness within each group (assuming willingness categories are "Yes", "No", "Unsure")
for content, group_data in listening_content_groups:
  willingness_counts = group_data['premium_sub_willingness'].value_counts()
  willingness_percentages = round((willingness_counts / willingness_counts.sum()) * 100, 1)  # Calculate percentages

  # Create a bar chart
  plt.figure(figsize=(6, 5))
  bars = willingness_counts.plot(kind='bar', color=willingness_colors)
  plt.xlabel("Premium Subscription Willingness")
  plt.ylabel("Count")
  plt.title(f"Premium Subscription Willingness for {content} Listeners")
  plt.xticks(rotation=0)  # Rotate x-axis labels for readability
  plt.yticks(fontsize=8)  # Adjust y-axis tick label size

  # Annotate bar labels with percentages (optional)
  for bar, perc in zip(bars.containers[0], willingness_percentages):
    plt.text(bar.get_x() + bar.get_width() / 2, perc, str(perc) + '%', ha='center', va='bottom', fontsize=8)

  plt.tight_layout()
  plt.show()
```
