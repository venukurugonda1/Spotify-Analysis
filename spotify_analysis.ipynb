```python
# Option 2: Impute missing values (e.g., with mode for categorical columns)
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
imputer = SimpleImputer(strategy='most_frequent', fill_value='unknown')
df[['preffered_premium_plan', 'fav_pod_genre', 'preffered_pod_format',
     'pod_host_preference', 'preffered_pod_duration']] = imputer.fit_transform(df[
         ['preffered_premium_plan', 'fav_pod_genre', 'preffered_pod_format',
          'pod_host_preference', 'preffered_pod_duration']])

# Check if 'Gender' is already categorical
if not pd.api.types.is_categorical_dtype(df['Gender']):
    categorical_columns = [col for col in df.columns if df[col].dtype == 'object']
else:
    categorical_columns = [col for col in df.columns if df[col].dtype == 'object' and col != 'Gender']

# Encode categorical columns with LabelEncoder
le = LabelEncoder()
for col in categorical_columns:
    df[col] = le.fit_transform(df[col])
```

```python
#Kmeans algorithem to cluster the data
# Feature selection for clustering (tailor based on your analysis)
from sklearn.cluster import KMeans
features = [ 'spotify_listening_device', 'fav_pod_genre', 'pod_lis_frequency']  # Consider adding more relevant features

# K-Means clustering with a predefined value for 'k' (or use the elbow method)
k = 3  # Replace with the optimal k from the elbow method or a suitable value

# Elbow Method (optional) to determine optimal number of clusters (k)
# ... (Implement elbow method code here to find the optimal 'k')

kmeans = KMeans(n_clusters=k)
kmeans.fit(df[features])

# Add cluster labels to the DataFrame
df['cluster_label'] = kmeans.labels_
```

```python
import matplotlib.pyplot as plt

# Scatter plot to visualize cluster separation based on chosen features
plt.figure(figsize=(8, 6))
plt.scatter(df[features[0]], df[features[1]], c=df['cluster_label'], cmap='viridis', alpha=0.7, s=50)
plt.xlabel(features[0])
plt.ylabel(features[1])
plt.title(f"K-Means Clustering (k={k}) - {features[0]} vs. {features[1]}")
plt.colorbar(label='Cluster Label')
plt.grid(True)
plt.show()
```

```python
# Bar plot to show the distribution of cluster sizes
plt.figure(figsize=(8, 6))
df['cluster_label'].value_counts().plot(kind='bar')
plt.title("Distribution of Cluster Sizes")
plt.xlabel("Cluster Label")
plt.ylabel("Count")
plt.show()
```

```python
# Heatmap to visualize the correlation between features
plt.figure(figsize=(8, 6))
corr_matrix = df[features].corr()
sns.heatmap(corr_matrix, annot=True, cmap='viridis')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def analyze_plan_usage_by_gender(data):
  """
  This function analyzes plan usage by gender in a provided dataset.

  Args:
      data: A pandas DataFrame containing columns named "Gender" and "spotify_subscription_plan".

  Returns:
      A dictionary where keys are genders and values are Series containing plan counts for each gender.
  """

  # Group data by gender
  gender_groups = data.groupby("Gender")

  # Analyze plan usage for each gender group
  plan_usage_by_gender = {}
  for gender, group_data in gender_groups:
    plan_counts = group_data["spotify_subscription_plan"].value_counts()
    plan_usage_by_gender[gender] = plan_counts

  return plan_usage_by_gender

# Example usage (assuming you have your data loaded into a DataFrame named 'df')
plan_usage_analysis = analyze_plan_usage_by_gender(df)
print("the predicted values:")
plan_usage_analysis


```

```python
# Extract data for visualization (assuming you retrieved data for each gender)
gender_labels = list(plan_usage_analysis.keys())  # Get gender labels
plan_counts_male = plan_usage_analysis[gender_labels[0]]  # Plan counts for first gender (modify index if needed)
plan_counts_female = plan_usage_analysis[gender_labels[1]]  # Plan counts for second gender (modify index if needed)

# Create colors for bars (customize as desired)
plan_colors = ['skyblue', 'lightgreen']

# Create a bar chart
plt.figure(figsize=(8, 6))
bar_width = 0.35  # Adjust bar width as needed
index = range(len(plan_counts_male))  # Create x-axis positions for bars

# Create bars for male and female plan usage
plt.bar(index, plan_counts_male, bar_width, label=gender_labels[0], color=plan_colors[0])
plt.bar([i + bar_width for i in index], plan_counts_female, bar_width, label=gender_labels[1], color=plan_colors[1])

# Label the chart
plt.xlabel("Subscription Plan")
plt.ylabel("Count")
plt.title("Spotify Subscription Plan Usage by Gender")
plt.xticks([i + bar_width / 2 for i in index], plan_counts_male.index, rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.legend()

# Display the bar chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have your data in a DataFrame named 'df'

# 1. Willingness vs. Preferred Premium Plan

# Create a cross-tabulation table
contingency_table = pd.crosstab(df['premium_sub_willingness'], df['preffered_premium_plan'])

# Get data for visualization (assuming 'Individual' and 'Family' are plan types)
willingness_categories = contingency_table.index  # Get willingness categories (e.g., "Yes", "No", "Unsure")
plan_types = contingency_table.columns  # Plan types (e.g., "Individual", "Family")
plan_counts_by_willingness = contingency_table.values  # Plan counts for each willingness category

# Create colors for bars (customize as desired)
willingness_colors = ['lightgreen', 'lightsteelblue', 'lightcoral']

# Create a stacked bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.35  # Adjust bar width as needed
index = range(len(plan_types))  # Create x-axis positions

# Stacked bars for plan counts within each willingness category
for i, willingness in enumerate(willingness_categories):
  plt.bar(index, plan_counts_by_willingness[i], bar_width, label=willingness, bottom=sum(plan_counts_by_willingness[:i]), color=willingness_colors[i])

# Label the chart
plt.xlabel("Premium Plan")
plt.ylabel("Count")
plt.title("Preferred Premium Plan by Spotify Subscription Willingness")
plt.xticks([i + bar_width / 2 for i in index], plan_types, rotation=45, ha='right')  # Rotate x-axis labels
plt.legend(title="Willingness")
plt.subplots_adjust(bottom=0.3)  # Adjust space for legend

# Display the stacked bar chart
plt.show()

# 2. Preferred Listening Content vs. Premium Willingness

# Group data by listening content
listening_content_groups = df.groupby('preferred_listening_content')

# Analyze willingness within each group (assuming willingness categories are "Yes", "No", "Unsure")
for content, group_data in listening_content_groups:
  willingness_counts = group_data['premium_sub_willingness'].value_counts()
  willingness_percentages = round((willingness_counts / willingness_counts.sum()) * 100, 1)  # Calculate percentages

  # Create a bar chart
  plt.figure(figsize=(6, 5))
  bars = willingness_counts.plot(kind='bar', color=willingness_colors)
  plt.xlabel("Premium Subscription Willingness")
  plt.ylabel("Count")
  plt.title(f"Premium Subscription Willingness for {content} Listeners")
  plt.xticks(rotation=0)  # Rotate x-axis labels for readability
  plt.yticks(fontsize=8)  # Adjust y-axis tick label size

  # Annotate bar labels with percentages (optional)
  for bar, perc in zip(bars.containers[0], willingness_percentages):
    plt.text(bar.get_x() + bar.get_width() / 2, perc, str(perc) + '%', ha='center', va='bottom', fontsize=8)

  plt.tight_layout()
  plt.show()
```
